### Hyperparameter Tuning Approach

To tune the hyperparameters, I chose to use the **RandomizedSearchCV** approach.

---

### Why RandomizedSearchCV?

Given the large number of hyperparameter combinations to evaluate, **RandomizedSearchCV** proved to be a more efficient choice.

---

### GridSearchCV vs. RandomizedSearchCV

The reason is straightforward:

- **GridSearchCV** is an exhaustive method that tests **every possible combination** of hyperparameters to find the optimal set. This can be very time-consuming.
- In contrast, **RandomizedSearchCV** samples a fixed number of **random combinations** from the parameter grid. This allows it to quickly and efficiently identify strong hyperparameter configurations with significantly **less computational effort**.

For the **RandomForestRegressor**, RandomizedSearchCV effectively selected the best parameter set after evaluating only a subset of all possible combinations.

---

### Performance Comparison

- **GridSearchCV**: Took **more than 10 minutes** to complete.
- **RandomizedSearchCV**: Completed in **just 2 minutes**, and also resulted in **better model performance**.
